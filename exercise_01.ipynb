```python
def process_string(s):
    reversed_str = s[::-1]
    every_second = s[1::2]

    print("Reversed string:", reversed_str)
    print("Every 2nd character:", every_second)

process_string("hello")
```

```python
def square_even(nums):
    result = [x**2 for x in nums if x % 2 == 0]
    print("Squares of even numbers:", result)

square_even([1, 2, 3, 4, 5, 6])
```

```python
def char_frequency(s):
    freq = {}
    for ch in s:
        freq[ch] = freq.get(ch, 0) + 1

    print("Character frequency:", freq)

char_frequency("banana")
```

```python
def odd_indices(lst):
    result = [i for i, val in enumerate(lst) if val % 2 != 0]
    print("Indices of odd numbers:", result)

odd_indices([1, 2, 3, 4, 5])
```

```python
def rotate_right(lst, k):
    k = k % len(lst)
    result = lst[-k:] + lst[:-k]
    print("Rotated list:", result)

rotate_right([1, 2, 3, 4, 5], 2)
```

```python
def lists_to_dict(keys, values):
    result = dict(zip(keys, values))
    print("Dictionary:", result)

lists_to_dict(['a', 'b', 'c'], [1, 2, 3])
```

```python
def manual_iteration(lst):
    it = iter(lst)
    print("Elements:")

    while True:
        try:
            item = next(it)
            print(item)
        except StopIteration:
            print("End of list")
            break

manual_iteration([10, 20, 30])
```

```python
def flatten_list(nested):
    result = [item for sublist in nested for item in sublist]
    print("Flattened list:", result)


flatten_list([[1, 2], [3, 4], [5]])
```

```python
def is_palindrome(s):
    result = s == s[::-1]
    print("Is palindrome:", result)


is_palindrome("madam")
```

```python
def second_largest(lst):
    first = second = float('-inf')

    for num in lst:
        if num > first:
            second = first
            first = num
        elif first > num > second:
            second = num

    print("Second largest number:", second)


second_largest([10, 20, 4, 45, 99])
```
